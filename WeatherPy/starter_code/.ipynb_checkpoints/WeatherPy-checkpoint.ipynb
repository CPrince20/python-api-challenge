```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and citiesz
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
country = []
lat = []
lng = []
wind = []
maxtemp = []
humidity = []
date = []
clouds = []

city_new = []

record = 0
record_set = 1

qurl = f"http://api.openweathermap.org/data/2.5/weather?appid={weather_api_key}&units=imperial&q="

for city in cities:
    try:
        reply = requests.get(qurl + city.replace(" ","&")).json()
        country.append(reply['sys']['country'])
        lat.append(reply['coord']['lat'])
        lng.append(reply['coord']['lon'])
        wind.append(reply['wind']['speed'])
        maxtemp.append(reply['main']['temp_max'])
        humidity.append(reply['main']['humidity'])
        date.append(reply['dt'])
        clouds.append(reply['clouds']['all'])
        
        
        if record > 48:
            city_new.append(city)
            record = 1
            record_set += 1
        
        else:
            city_new.append(city)
            record +=1
            
        print(f"Gathered Record {record} of Set {record_set} / {city}")
            
    except Exception:
              print("Data not found")
              
print("Task Executed!")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
dictionary = {
    "Country": country,
    "City": city_new,
    "Lat": lat,
    "Lng": lng,
    "Wind Level": wind,
    "Max Temperature": maxtemp,
    "Humidity": humidity,
    "Date": date,
    "Cloudiness": clouds
}
```

```python
weatherdf = pd.DataFrame(dictionary)

weatherdf = weatherdf.dropna()
weatherdf.head()
```

```python
weatherdf.to_csv("../output_new/weatherdata.csv", index=False, header=True)
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
```

```python
# Create Northern and Southern Hemisphere DataFrames

north = []
south = []

for ind in weatherdf.index:
    if weatherdf['Lat'].loc[ind] >= 0:
        north.append(weatherdf.loc[ind])
    else:
        south.append(weatherdf.loc[ind])
```

```python
northdf = pd.DataFrame(north)
northdf.head()
southdf = pd.DataFrame(south)
southdf.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x = []
x = northdf['Max Temperature']

y = []
y = northdf['Lat']
slope, intercept, r_value, p_value, std_err = linregress(x, y)

plt.plot(x, y, 'o', label='North Hempiphere Max Temp vs Lat')
plt.plot(x, intercept + slope*x, 'r', label='fitted line')
plt.legend()
plt.show

plt.savefig("../output_new/NorthLatvTemp.png")


### This shows there is a strong linear relationship between the
### Northern Hemisphere's Maximum Temperature and Latitude
```
